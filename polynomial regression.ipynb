```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(500)
pageSpeeds= np.random.normal(3.0,1.0,100)
purchaseAmount=np.random.normal(50.0,30.0,100)/ pageSpeeds
```

```python
plt.scatter(pageSpeeds,purchaseAmount,color='purple')
plt.show()
```

```python
#now we will split the data into train and test- 80% train and 20% test to avoid overfitting
trainX=pageSpeeds[:80]
testX=pageSpeeds[80:]

trainY=purchaseAmount[:80]
testY=purchaseAmount[80:]
```

```python
plt.scatter(trainX,trainY,color='purple')
plt.show()
```

```python
plt.scatter(testX,testY,color='red')
plt.show()
```

```python
#we will fit 8th degree polynomial to this data (which is almost certainly overfitting)

X=np.array(trainX)
Y=np.array(trainY)
p4=np.poly1d(np.polyfit(X,Y,8))

#later on experiment with reducing the degree and check for model performance
```

```python
xp=np.linspace(start=0,stop=7,num=100)
axes=plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,100])
plt.scatter(X,Y)
plt.plot(xp,p4(xp),c='r')
plt.show()

#overfitting resulted
```

```python
#lets do this on our test data

testX=np.array(testX)
testY=np.array(testY)

axes=plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,100])
plt.scatter(testX,testY)
plt.plot(xp,p4(xp),c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2= r2_score(testY,p4(testX))  #check model performance on the test data
print(r2)
```

```python
#test model performance on the training data

r2_train=r2_score(trainY,p4(trainX))
print (r2_train)
```
