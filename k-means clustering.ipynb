```python
#create fake income/ age data for a said no. of people for a said no. of clusters
import random
import numpy as np
def cluster_data_gen(N, k):
    random.seed(10)
    people_per_cluster=float(N)/k
    X=[] #create an empty array
    for i in range (k):
        income_centroid=random.uniform(20000.0,30000.0)
        age_centroid=random.uniform(20.0,70.0)
        for j in range(int(people_per_cluster)):
            X.append([np.random.normal(income_centroid,10000.0),np.random.normal(age_centroid,2.0)])
    X=np.array(X)
    return X
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
data=cluster_data_gen(100,5)
```

```python
model=KMeans(n_clusters=5) #picking k as 5

#note I am scaling the data for normalizing it! It is important for good results

```

```python
#We can look at clusters each data point got assigned to
print (model.labels_)
```

```python
model=model.fit(scale(data)) #normalizing the data along with fitting the model

#We will visualize it 
plt.figure(figsize=(8,6))
plt.scatter(data[:,0],data[:,1],c=model.labels_.astype(np.float))
plt.show()
```
